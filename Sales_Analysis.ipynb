```python
import pandas as pd
import os
```

```python
files = [file for file in os.listdir('./Sales_Data')]
for file in files:
    print(file)
files
```

```python
mergedAll = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    mergedAll = pd.concat([mergedAll,df])
mergedAll.to_csv('df.csv',index=False)
```

```python
df = pd.read_csv('df.csv')
df.head()
```

```python
df['Month'] = df['Order Date'].str[:2]
df
```

```python
df['Month'].dtypes
```

```python
df = df.loc[df['Month'] != 'Or']
df
```

```python
df = df.dropna(how='all')
df
```

```python
df['Month'] = df['Month'].astype('int32')
df
```

```python
df.dtypes
```

```python
df[['Price Each','Quantity Ordered']].dtypes
```

```python
df['Sales'] = df['Price Each']*df['Quantity Ordered']
df.head()
```

```python
df['Price Each'] = pd.to_numeric(df['Price Each'])
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
```

```python
df['No of Orders'] = 1
df
#df.groupby('Month').count()['Count']
```

```python
dfc = df.groupby(['Month']).sum()
dfc[['Sales','Quantity Ordered', 'No of Orders']].sort_values(['Sales'], ascending = False)
```

```python
result = df.groupby(['Month']).sum()
result
```

```python
import matplotlib.pyplot as plt
```

```python
months = range(1, 13)
name = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.bar(months, result['Sales'], color ='purple')
plt.xlabel('Month Number', color='white')
plt.ylabel('Sales in $USD', color='white')
plt.xticks(months, name, color='white')
plt.yticks(color ='white')
plt.show()
```

```python
def get_city(address):
    return address.split(', ')[1]

def get_state(address):
    return address.split(', ')[2].split( )[0]

df['City'] = df['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})" )
df
```

```python
results = df.groupby('City').sum()
results
```

```python
import matplotlib.pyplot as plt
cities = df['City'].unique()
```

```python
cities = [city for city, df in df.groupby('City')]
print(cities)

plt.bar(cities, results['Sales'], color='maroon')
plt.xticks(cities, rotation=45,color='white', size=9)
plt.yticks(color='white')
plt.xlabel("Month number", color='white')
plt.ylabel("Sales in USD($)", color='white')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Hour'] = df['Order Date'].dt.hour
df['Mins'] = df['Order Date'].dt.minute
df['Year'] = df['Order Date'].dt.year
df['Months'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day
#df = df.drop(columns=['Secs'])
df
```

```python
df.head()
```

```python
no_of_orders = df.groupby('Hour').sum()['No of Orders']
```

```python
import matplotlib.pyplot as plt 
hours = [hour for hour, df in df.groupby('Hour')]
plt.plot(hours, no_of_orders)
plt.grid()
plt.xticks(hours)
plt.xlabel('Hours')
plt.ylabel('No. of Orders')
plt.show()
```

```python
newD = df.loc[df['Order ID'].duplicated(keep=False)]
newD.head(5)
```

```python
newD['Grouped'] = newD.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
newD.head(5)
#newD = newD.drop(columns=['Months'])
```

```python
newD = newD[['Order ID', 'Grouped']].drop_duplicates()
newD
```

```python
prod_g = df.groupby('Product')
quantity_ordered = prod_g.sum()['Quantity Ordered']

products = [product for product, df in prod_g]

plt.bar(products, quantity_ordered)
plt.xlabel('Product', color='white')
plt.ylabel('Quantity Ordered', color='white')
plt.xticks(products, rotation=90, size=9, color='white')
plt.yticks(color='white')
plt.show()
```

```python
prices = df.groupby('Product').mean()['Price Each']
prices
```

```python


fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product Name', color='white')
ax1.set_ylabel('Quantity Ordered', color='white')
ax2.set_ylabel('Price($)', color='white')
ax1.set_xticklabels(products, rotation=90, size=9, color='white')
plt.show()
```
